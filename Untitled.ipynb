```python
#lets turn the university attended column into either university, polytechnic or others - 1,0,2
university = []
for item in container[5]:
    if "university" in item.lower():
        university.append(1)
    elif "polytechnic" in item.lower():
        university.append(0)
    else: university.append(2) #other federal institute 

university = pd.DataFrame(university)
university.shape
```

```python
#lets turn the courses into integer from 1-10 based on top 10 courses and 11 for others; 1 - 10
courses = []
for item in container[4]:
    if "computer science" in item.lower():
        courses.append(1)
    elif "accountancy" in item.lower():
        courses.append(2)
    elif "economics" in item.lower():
        courses.append(3)
    elif "electrical & electronic engineering" in item.lower():
        courses.append(4)
    elif "mass communication (communication and language arts)" in item.lower():
        courses.append(5)
    elif "business administration" in item.lower():
        courses.append(6)
    elif "biochemistry" in item.lower():
        courses.append(7)
    elif "microbiology" in item.lower():
        courses.append(8)
    elif "mechanical engineering" in item.lower():
        courses.append(9)
    elif "english and literary studies" in item.lower():
        courses.append(10)
    else: courses.append(11) #where 11 is others
    
courses = pd.DataFrame(courses)
courses.shape
```

```python
#lets turn the sex to either male or female; 1 or 0
sex = []
for item in container[2]:
    if "male" in item.lower():
        sex.append(1)
    elif "female" in item.lower():
        sex.append(0)
    else:sex.append(2) #others
sex = pd.DataFrame(sex)
sex.shape
```

```python
#lets see our current status column
sr
```

```python
#lets turn the current status to either employed or unemployed; 1 or 0
status = []
for item in container[7]:
    if "working full time (paid employment)" in item.lower():
        status.append(1)
    elif "unemployed" in item.lower():
        status.append(0)
    else:status.append(0) #we are only after those employed and unemployed; we believe the rest are either unemployed or became self employed out of unemployement
status = pd.DataFrame(status)
status.shape
```

```python
#lets drop the 1 NaN that will obstruct our loop and replace it with 0 later
container[9].dropna(axis=0, inplace=True)
```

```python
#lets turn the completed NYSC? to either yes or no; 1 or 0
nysc = []
for item in container[9]:
    if "yes" in item.lower():
        nysc.append(1)
    elif "no" in item.lower():
        nysc.append(0)
    else:nysc.append(0) #ongoing and non recognised attributes will be recorded as void
nysc.append(0) #for the NaN that was dropped
nysc = pd.DataFrame(nysc)
nysc.shape
```

```python
container.shape
```

```python
#lets now create our new dataframe of all features
predictingDF = pd.concat([university,courses,sex,nysc,status], axis=1)
columns = "University or Poly?", "Course studied in the Top 10", "Sex", "Completed NYSC?", "Current Status"
#lets change the columns of our new dataframe
predictingDF.columns = columns
predictingDF.head()
```

```python
#since we are dealing with numbers, lets normalize our data frame so it can be well understood by the model
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(predictingDF.drop('Current Status',axis=1))
X = scaler.transform(predictingDF.drop('Current Status',axis=1))
X
```

```python
X.shape
```

```python
y = predictingDF["Current Status"]
y.shape
```

```python
#lets now split our data into training and testing set before instantiating our model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# our data has been splitted into 80% training and 20% testing sets
# lets now instatiate our model (logistic regression ) and train our samples
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=4.0, solver="newton-cg")
#where C is the inverse of regularization and solver is our optimization algorithm
```

```python
# lets now train the model
lr.fit(X_train, y_train)
```

```python
# wonderful! our model has been trained to make predictions!!
# lets make predicitions based on the remaining 20% and evaluate to see how well our model performed
predictions = lr.predict(X_test)
predictions
```

```python
# Our model has made some wonderful predicitions, lets build an evaluation to evaluate the efficiency of this model
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
